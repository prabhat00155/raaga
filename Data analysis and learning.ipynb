```python
from audioFeatureExtraction import loadDataset
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
X = []
Y = []
for row in loadDataset():
    for target, feature in row.items():
        X.append(feature)
        Y.append(int(target == 'mohanam'))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
SGDClf = SGDClassifier(max_iter=10000, tol=1e-5, random_state=0)
pca = PCA()
pipe = Pipeline(steps=[('pca', pca), ('logistic', SGDClf)])
param_grid = {
    'pca__n_components': [2, 6, 12, 24, 36],
    'logistic__alpha': np.logspace(-6,-4, 4, 5),
    'logistic__loss':['hinge', 'log', 'modified_huber', 'perceptron'],
    'logistic__penalty':['l2', 'l1', 'elasticnet']
    }
SGDsearch = GridSearchCV(pipe, param_grid, iid=False, cv=5)
SGDsearch.fit(X_train, y_train)
```

```python
print(SGDsearch.best_score_)
```

```python
print(SGDsearch.best_params_)
```

```python
results = pd.DataFrame(SGDsearch.cv_results_)
```

```python
results
```

```python
clf = GradientBoostingClassifier()
pipe = Pipeline(steps = [('GBC', clf)])
param_grid = {
        'GBC__loss':["deviance", "exponential"],
        'GBC__learning_rate':[0.1, 0.001, 1, 0.0001],
        'GBC__n_estimators':[100, 1000, 500, 50,2000],
        'GBC__max_depth':[3,6,12],
        'GBC__max_features':["auto", "sqrt", "log2", None],
    }
GBCsearch = GridSearchCV(pipe, param_grid, iid=False, cv=5)
GBCsearch.fit(X_train,y_train)
```

```python
GBCsearch.best_score_
```

```python
GBCsearch.best_params_
```

```python
GBCsearch.score(X_test, y_test)
```
